```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import scale

data = pd.read_csv('/content/drive/MyDrive/wine.data',
                   index_col=None,
                   header=None,
                   names=['class', 'Alcohol', 'Malic acid', 'Ash',
                          'Alcalinity of ash',
                          'Magnesium', 'Total phenols', 'Flavanoids',
                          'Nonflavanoid phenols','Proanthocyanins',
                          'Color intensity', 'Hue',
                          'OD280/OD315 of diluted wine','Proline'])
data
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
X = data[['Alcohol', 'Malic acid', 'Ash',
                          'Alcalinity of ash',
                          'Magnesium', 'Total phenols', 'Flavanoids',
                          'Nonflavanoid phenols','Proanthocyanins',
                          'Color intensity', 'Hue',
                          'OD280/OD315 of diluted wine','Proline']]
y = data['class']
cv_accuracy = [cross_val_score(estimator=KNeighborsClassifier(n_neighbors=k),
                               X=X, y=y,
                               cv=kf).mean() for k in range(1,51)]
max_acc = max(cv_accuracy)
print(cv_accuracy.index(max_acc)+1)
print(max_acc)
```

```python
X_scaled = X.copy()
X_scaled = scale(X_scaled)
```

```python
cv_accuracy_scaled = [cross_val_score(estimator=KNeighborsClassifier(n_neighbors=k),
                               X=X_scaled, y=y,
                               cv=kf).mean() for k in range(1,51)]
max_acc_s = max(cv_accuracy_scaled)
print(cv_accuracy_scaled.index(max_acc_s)+1)
print(max_acc_s)
```
